```python
#Hypothesis
#H0:p1=p2 (females having hypertension in basti are same as in chitrakoot)
#Ha:p1>p2 (females having hypertension in basti are significantly greater than in chitrakoot)

significance = 0.05

Females_Hypertension_basti, N1 = (753, 3765)
Females_Hypertension_chitrakoot, N2 = (284, 3639)

successes = np.array([Females_Hypertension_basti, Females_Hypertension_chitrakoot])
samples = np.array([N1, N2])

stat, p_value = proportions_ztest(count=successes, nobs=samples,  alternative='larger')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

if p_value > significance:
   print ("Fail to reject the null hypothesis -females having hypertension in basti are same as in chitrakoot")
else:
   print ("Reject the null hypothesis - females having hypertension in basti are significantly greater than in chitrakoot")
```

```python
from statsmodels.stats.proportion import proportions_ztest
#Hypothesis
#H0:P = 0.258 (There is no significance difference in prevalence of male hypertension in 2015 and prevalence of male Hypertension in 2010.)
#Ha:P < 0.258 (Prevalence of Male hypertension in 2015 is significantly less than prevalence of male Hypertension in 2010.)
significance = 0.05

sample_success = 573
sample_size = 2561

null_hypothesis = 0.258

stat, p_value = proportions_ztest(count=sample_success, nobs=sample_size, value=null_hypothesis, alternative='smaller')

# report
print('z_stat: %0.3f, p_value: %0.3f' % (stat, p_value))

if p_value > significance:
   print ("Fail to reject the null hypothesis - There is no significance difference in prevalence of male hypertension in 2015 and prevalence of male Hypertension in 2010")
else:
   print ("Reject the null hypothesis -Prevalence of Male hypertension in 2015 is significantly less than prevalence of male Hypertension in 2010.")
```
